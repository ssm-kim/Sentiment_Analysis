# 형태소 분석

```python
from konlpy.tag import Mecab, Komoran
tokenizer = Komoran()
```

```python
token = Komoran(userdic="./user_dict.txt")
```

```python
df_dgbdata.iloc[0,0]
```

```python
tokenizer.pos(df_dgbdata.iloc[0,0])
```

```python
token.pos(df_dgbdata.iloc[0,0])
```

```python
word_dict ={}
for cont in df_dgbdata.title:
    for word, morpheme in token.pos(cont):
        if morpheme in ['NNG', 'NNP', 'NNB', "NNM"] and len(word) > 1 : #NNG NNP NNB NNM
            if word_dict.get(word) == None:
                word_dict[word] = 1
            else:
                word_dict[word] += 1
```

```python
sorted(word_dict.items(), key=lambda x : x[1], reverse=True)
```

# 가장 많이 쓰이는 "명사" 알아보기

```python
f = open("GoogleNews_DGB.csv", encoding="utf-8")

sample_dict = {}
line_number = 0

for line in f:
    # 처리중 라인 카운트 (optional)
    #line_number += 1
    #print(line_number, end="\r")
    
    # 여기에 코드를 작성합니다.
    line = line.strip()
    word_list = line.split()
    for word in word_list:
        if word in sample_dict.keys():
            sample_dict[word] += 1
        else:
            sample_dict[word] = 1
    #
    
f.close()

sample_list = []

for sample in sample_dict.items():
    sample_list.append(sample)

sample_list.sort(key=lambda elem: elem[1], reverse=True)
sample_list[:100]
```
